```python
df = pd.read_csv('./서울시 지하철 호선별 역별 시간대별 승하차 인원 정보.csv'
                , encoding='cp949')
```

```python
df['사용월'] = df['사용월'].astype('str')
df.dtypes
```

```python
df.drop(columns=['작업일자'],inplace=True)
```

```python
#공통컬럼
df1 = df.iloc[:,:3]
#승차컬럼
df2 = df.iloc[:,3::2]
#하차컬럼
df3 = df.iloc[:,4::2]
```

```python
#승차테이블
df2.columns = df2.columns.str.split(' ').str[0] #['04시-05시','승차인원']중 앞에값사용
df_in = pd.concat([df1,df2],axis=1)
df_in
```

```python
# 하차테이블
df3.columns = df3.columns.str.split(' ').str[0]
df_out = pd.concat([df1,df3],axis=1)
df_out
```

최근월을 기준으로 승하차 데이터 프레임생성

```python
df_in_202108 = df_in[df_in['사용월']=='202108']
df_out_202108 = df_out[df_out['사용월']=='202108']
```

```python
# 출근시간에 가장 많은 사람이 승차하는 역은 어디?(08-09)
df_in_202108.nlargest(10, '08시-09시')[['지하철역','08시-09시']]
```

```python
# 출근시간에 가장 많은 사람이 하차하는 역은 어디?(09-10)
df_out_202108.nlargest(10,'09시-10시')[['지하철역','09시-10시']]
```

```python
# 퇴근시간에 가장 많은 사람이 승차하는 역은 어디?(18-19)
df_in_202108.nlargest(10,'18시-19시')[['지하철역','18시-19시']]
```

```python
# 퇴근시간에 가장 많은 사람이 하차하는 역은 어디?(19-20)
df_in_202108.nlargest(10,'19시-20시')[['지하철역','19시-20시']]
```

```python
# 강남역 최근 시간대별 차 정보 분석
df_gangnam_in = df_in_202108[df_in_202108['지하철역']=='강남'].iloc[:,3:]
```

```python
df_gangnam_in = df_gangnam_in.melt()
```

```python
#컬럼명 변경
df_gangnam_in.columns=['시간대','승차건수']
df_gangnam_in.sort_values('승차건수',ascending=False)
```

```python
#시간대별 승차인원 시각화 하기
plt.figure(figsize=(10,7))
plt.barh(df_gangnam_in['시간대'],df_gangnam_in['승차건수'])
plt.show()
```
